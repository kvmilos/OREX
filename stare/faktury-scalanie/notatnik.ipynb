```python
import pandas as pd
import numpy as np
```

### DO DZIALU ROZLICZEN

```python
lista = pd.read_excel('lista.xlsx')
naglowki = pd.read_excel('naglowki.xlsx')
wiersze = pd.read_excel('wiersze.xlsx')
```

```python
lista.head()
```

```python
naglowki.head()
```

```python
wiersze.head()
```

```python
# left join naglowki and wiersze on naglowki: 'Nr' and wiersze: 'Nr dokumentu' 
df = pd.merge(naglowki, wiersze, left_on='Nr', right_on='Nr dokumentu', how='left')
df.head()
df.drop(columns=['Nr odbiorcy płatności', 'Nr dokumentu', 'Nr wiersza', 'Nr dostawcy (zakup)_y', 'Nr_y', 'Opis', 'Opis 2', 'Kwota', 'Kwota z VAT'], inplace=True)
df.head()
```

```python
lista
```

```python
df = df.groupby(['Nr faktury dostawcy', 'Nr dostawcy (zakup)_x']).agg({'Nr_x': lambda x: list(x.unique()), 'Reservation No.': lambda x: [int(i) for i in x if not pd.isna(i)]})
df.head()
```

```python
# right merge df and lista on: df: 'Nr faktury dostawcy' with lista: 'Nr dokumentu zewnętrznego' and df: 'Nr dostawcy (zakup)_x' with lista: 'Nr dostawcy'
df = pd.merge(df, lista, left_on=['Nr faktury dostawcy', 'Nr dostawcy (zakup)_x'], right_on=['Nr dokumentu zewnętrznego', 'Nr dostawcy'], how='right')
df.head()
```

```python
# to excel
df.to_excel('output.xlsx', index=False)
```
