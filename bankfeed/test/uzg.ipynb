```python
import pandas as pd
```

```python
obroty = pd.read_excel('uzgodnienie.xlsx', sheet_name='obroty', header=None)
ma = pd.read_excel('uzgodnienie.xlsx', sheet_name='ma', header=0)
obroty.columns = ['nr dostawcy', 'nazwa', 'kwota wn', 'kwota ma']
```

```python
obroty
```

```python
ma
```

```python
# change obroty in a way that groups the dataframe by 'nr dostawcy', 'nazwa' and sums the 'kwota wn' and 'kwota ma' columns
obroty = obroty.groupby(['nr dostawcy', 'nazwa']).sum().reset_index()
obroty
```

```python
# change ma in a way that groups the dataframe by 'nr dostawcy w symfonii' and sums the 'Kwota Ma' column, and takes care of all other columns
ma = ma[['nr dostawcy w symfonii', 'Kwota Ma']]
ma = ma.groupby('nr dostawcy w symfonii').sum().reset_index()
ma
```

```python
# do an outer join of them both, on the 'nr dostawcy' and 'nr dostawcy w symfonii' columns respectively, summing the 'kwota ma' and 'Kwota Ma' columns from dataframes respectively, but not together
result = pd.merge(obroty, ma, left_on='nr dostawcy', right_on='nr dostawcy w symfonii', how='outer')
result['kwota ma'] = result['kwota ma'].fillna(0)
result['Kwota Ma'] = result['Kwota Ma'].fillna(0)
result['nr dostawcy'] = result['nr dostawcy'].fillna(result['nr dostawcy w symfonii'])
result.fillna(0, inplace=True)
result = result[['nr dostawcy', 'nazwa', 'kwota ma', 'Kwota Ma']]
result['diff'] = result['kwota ma'] - result['Kwota Ma']
```

```python
result
```

```python
# save to excel file
result.to_excel('result.xlsx', index=False)
```
