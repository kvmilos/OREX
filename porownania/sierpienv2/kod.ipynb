```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/kvmilos/OREX/main/kontrahenci.csv'
df = pd.read_csv(url, dtype=str)
dic = dict(zip('201-2-1-' + df['pozycja'], df['kod']))

def get_reservation(x: str):
    if x in dic:
        return dic[x]
    else:
        return x
```

```python
samo = pd.read_excel('sierpien.xlsx', sheet_name='samo', header=0)
```

```python
samo.head()
```

```python
samo['reservation'] = samo['reservation'].astype(str)
```

```python
samo.head()
```

```python
df703 = pd.read_excel('sierpien.xlsx', sheet_name='703', header=None)
df704 = pd.read_excel('sierpien.xlsx', sheet_name='704', header=None)
df763 = pd.read_excel('sierpien.xlsx', sheet_name='763-7', header=None)

df703['source'] = '703'
df704['source'] = '704'
df763['source'] = '763-7'
```

```python
df703.columns = ['c1', 'c2', 'c3', 'c4', 'data', 'dokument', 'opis', 'c8', 'kwota', 'konto', 'c10', 'zrodlo']
df704.columns = ['c1', 'c2', 'c3', 'c4', 'data', 'dokument', 'opis', 'c8', 'kwota', 'konto', 'c10', 'zrodlo']
df763.columns = ['c1', 'c2', 'c3', 'c4', 'data', 'dokument', 'opis', 'c8', 'kwota', 'konto', 'c10', 'zrodlo']
```

```python
symfonia = pd.concat([df703, df704, df763], ignore_index=True)
symfonia['rezerwacja'] = symfonia['konto'].apply(get_reservation)
```

```python
symfonia['konto'] = symfonia['konto'].astype(str)
```

```python
symfonia.head()
```

```python
# save it to excel 'slownik':
df = symfonia[symfonia['rezerwacja'].str.len() > 7]['rezerwacja'].value_counts()
df.to_excel('slownik.xlsx')
```

```python
symfonia2 = symfonia.copy()
```

```python
how = pd.read_excel('slownik.xlsx', header=None)
```

```python
how.columns = ['from', 'to']
how['from'] = how['from'].astype(str)
```

```python
def change_reservation(x: str, how):
    if x in how['from'].values:
        return str(how[how['from'] == x]['to'].values[0])
    else:
        return x

symfonia2['rezerwacja'] = symfonia2['rezerwacja'].apply(lambda x: change_reservation(x, how) if x in how['from'].values else x)
```

```python
symfonia2.head()
```

```python
symfonia2[symfonia2['rezerwacja'].str.len() > 7]['rezerwacja'].unique()
```

```python
symfonia2['rezerwacja'] = symfonia2['rezerwacja'].astype(str)
```

```python
result = pd.DataFrame(columns=['res', 'sum_symfonia', 'SAMO_Cat_Price', 'difference'])
result['res'] = pd.concat([symfonia2['rezerwacja'], samo['reservation']]).unique()
result['sum_symfonia'] = result['res'].apply(lambda x: symfonia2[symfonia2['rezerwacja'] == x]['kwota'].astype(float).sum())
result['SAMO_Cat_Price'] = result['res'].apply(lambda x: samo[samo['reservation'] == x]['Catalogue price'].astype(float).sum())
result['difference'] = (result['sum_symfonia'] - result['SAMO_Cat_Price']).round(2)
result['abs_difference'] = result['difference'].abs()
result['partner'] = result['res'].apply(lambda x: samo[samo['reservation'] == x]['Partner'].unique()[0] if len(samo[samo['reservation'] == x]['Partner'].unique()) > 0 else '')
result['zrodlo'] = result['res'].apply(lambda x: symfonia2[symfonia2['rezerwacja'] == x]['zrodlo'].unique()[0] if len(symfonia2[symfonia2['rezerwacja'] == x]['zrodlo'].unique()) > 0 else '')
```
